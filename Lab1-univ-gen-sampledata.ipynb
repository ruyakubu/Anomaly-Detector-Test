```python
import os
import time
import uuid
import json
import random
import datetime
import time
import numpy as np
from azure.iot.device import IoTHubDeviceClient, Message

# The connection string for a device should never be stored in code. For the sake of simplicity we're using an environment variable here.
conn_str = os.getenv("IOTHUB_DEVICE_CONNECTION_STRING")

# The client object is used to interact with your Azure IoT hub.
device_client = IoTHubDeviceClient.create_from_connection_string(conn_str)

num_of_records = 60

# Connect the client.
device_client.connect()

# send telemetry data to IoT Hub
def send_message_to_cloud(data):
    print("sending message to IoT Hub")
    
    #convert data to json format
    json_obj = json.dumps(data, indent = 2)
    msg = Message(json_obj)
    msg.message_id = uuid.uuid4()
    msg.content_encoding = "utf-8"
    msg.content_type = "application/json"
    device_client.send_message(msg)
    
#simulate sensor data    
def generate_data():
    
    entry_list = list()
    i = 0
    
    while i < 60:
        num = random.randint(25, 70)
        print("Temperature reading: ", num)
        #datetime.fromtimestamp(timestamp, timezone.utc)
        date = datetime.datetime.now().strftime("%Y-%m-%dT%H:%M:00Z")        
        entry = {}
        entry['timestamp'] = date
        entry['value'] = num
        
        #send data to IoT Hub
        send_message_to_cloud(entry)
        entry_list = np.append(entry_list, entry).tolist()
        
        i = i + 1
        time.sleep(60)
   
    #send json file to local directory
    resquest_data = {}
    resquest_data['granularity'] =  "minutely"
    resquest_data['series'] = entry_list 
    
    #generate json file
    json_object = json.dumps(resquest_data, indent = 2)
    print("json file: " + json_object)
    # Writing to file
    with open("sample_data.json", "w") as outfile:
        outfile.write(json_object)
            
generate_data()

# finally, shut down the client
device_client.shutdown()
```
