```python
import csv
import time
import uuid
import json
import random
import datetime
import time
import zipfile
import numpy as np


ANOMALY_INDEXES = [12, 25]

#simulate sensor data    
def generate_senor_data():  
    hf = open("humidity.csv", "a", newline="")
    tf = open("temperature.csv", "a", newline="")
    
    header = ("timestamp","value")
   
    humidity_writer = csv.writer(hf)   
    temp_writer = csv.writer(tf)
    
    humidity_writer.writerow(header)
    temp_writer.writerow(header)
    
    date = datetime.datetime.now()
    
    i = 0
    while i < 3000:
        print("counter: ", i)
        humid = random.randint(40, 60)
        temp = random.randint(25, 70)
        date = date + datetime.timedelta(minutes = 1)
        
        dt = date.strftime("%Y-%m-%dT%H:%M:00Z") 
        
        tentry_list = list()
        tentry_list = np.append(tentry_list, dt).tolist()
        tentry_list = np.append(tentry_list, temp).tolist()
        tentry = tuple(tentry_list)
        
        hentry_list = list()
        hentry_list = np.append(hentry_list, dt).tolist()
        hentry_list = np.append(hentry_list, humid).tolist()
        hentry = tuple(hentry_list)
        
        temp_writer.writerow(tentry) 
        humidity_writer.writerow(hentry)    
        i = i + 1
        #time.sleep(60)
        
    tf.close()
    hf.close()
    

def zip_data_files(): 
    
    generate_senor_data()
       
    data_files = ['temperature.csv', 'humidity.csv']  
    with zipfile.ZipFile('weather_data.zip', 'w') as zipF:
        for file in data_files:
            zipF.write(file, compress_type=zipfile.ZIP_DEFLATED)
    print('Data files has been compressed')
    
    
zip_data_files()
```
